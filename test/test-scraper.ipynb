```python
file_path = f"data/incessantbeat_authors_metadata.txt"
with open(file_path) as f:
    meta_dict = json.load(f)
df = pd.DataFrame(meta_dict)
df
```

```python
df['year'] = pd.to_datetime(df['published']).dt.year
df['chaps'] = df['chapters'].str.split('/').str[0]
for c, w in zip(['comments', 'kudos', 'bookmarks', 'hits'], [1.5,.5,1,0.1]):
    df[c] = df[c].astype(float)
    df[f'avg_{c}'] = df[c] / df['chaps'].astype(float)
    df[f'weighted_{c}'] = df[f'avg_{c}'] * w
df['score'] = df[['weighted_comments', 'weighted_kudos', 'weighted_bookmarks', 'weighted_hits']].mean(axis=1) / (2022.1 - df['year'])
```
